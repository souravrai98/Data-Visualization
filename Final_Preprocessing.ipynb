```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Dataset is loaded

```python
album = pd.read_csv("wasabi_albums.csv")
```

```python
album.head()
```

Frequency per genre per country is calculated and the top genre per country is selected

```python
listc = list(album["country"].unique())
famous = pd.DataFrame(columns=["country","genre","Freq"])
for i in listc:
    if not album[album["country"]==i]['genre'].value_counts().index.empty:
       famous = famous.append({"country":i,"genre":album[album["country"]==i]['genre'].value_counts().index[0],"Freq":album[album["country"]==i]['genre'].value_counts().max()},ignore_index=True)
       #print("For country ",i," the top genre is ",album[album["country"]==i]['genre'].value_counts().index[0] )
```

```python
famous.head()
```

Full name of the country is added

```python
country_code = pd.read_csv("wikipedia-iso-country-codes.csv")
```

```python
country_code.head()
```

```python
for i in famous["country"]:
    if country_code[country_code['Alpha-2 code'] == i]['English short name lower case'].values.size != 0 :
         famous.loc[famous["country"]==i,"Full_name"]= country_code[country_code['Alpha-2 code'] == i]['English short name lower case'].values[0]
    
```

```python
famous.head()
```

```python
famous_cleaned = famous.dropna()
```

```python
famous_cleaned
```

Latitude and Longitude is added

```python
df = pd.read_csv("countries.csv")
```

```python
df.head()
```

```python
famous_cleaned["longitude"] = np.nan
famous_cleaned["latitude"] = np.nan
```

```python
for i in famous_cleaned["country"]:
    if df[df['country'] == i]['longitude'].values.size != 0:
        famous_cleaned.loc[famous_cleaned["country"]==i,"longitude"] = df[df['country'] == i]['longitude'].values[0]
        famous_cleaned.loc[famous_cleaned["country"]==i,"latitude"] = df[df['country'] == i]['latitude'].values[0]
```

```python
famous_cleaned.loc[famous_cleaned["country"]=="UM","longitude"] = -159.996719
famous_cleaned.loc[famous_cleaned["country"]=="UM","latitude"] = -0.374350
```

```python
famous_cleaned[famous_cleaned.isna().any(axis=1)]
```

Group column is added

```python
famous_cleaned["group"] = np.nan
```

```python
def conditions(s):
    if (s['genre'] == "Rock"):
        return "A"
    if (s["genre"] == "Pop"):
        return "B"
    if (s["genre"] == "Alternative Rock"):
        return "C"
    if (s["genre"] == "Heavy Metal"):
        return "D"
    if (s["genre"] == "Indie Pop"):
        return "E"
    if (s["genre"] == "Punk Rock"):
        return "F"
    if (s["genre"] == "Black Metal"):
        return "G"
    else:
        return "H"
```

```python
famous_cleaned['group'] = famous_cleaned.apply(conditions, axis=1)
```

```python
famous_cleaned.head()
```

Final csv is generated

```python
famous_cleaned.to_csv('genre_final.csv')
```
